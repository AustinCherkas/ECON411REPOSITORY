# Question 2

use the variance function the textbook to find the variance of this list. Assume the list represents a population in whole.

```python
def total(values):
    total_ = 0
    for val in randomlist1:
        total_ += val
    return total_
total(values)
```

```python
def mean(randomlist1):
    n = len(randomlist1)
    mean1 = total(randomlist1) / n
    return mean1
mean(randomlist1)
```

```python
def variance(randomlist1, sample = True):
    listmean = mean(randomlist1)
    n = len(randomlist1)
    DOF = n - 1
    sumsquarediff = 0
    
    for val in randomlist1:
        diff = val - listmean
        sumsquarediff += (diff) ** 2
    if sample == False:
        variance1 = sumsquarediff / n
    else:
        variance1 = sumsquarediff / DOF
    return variance1
        
variance(randomlist1, sample = True), variance(randomlist1, sample = False)
```

# Question 6

Explain why it might be advantageous to create a function instead of building all commands from scratch as you create a script.

Functions help reduce the amount of code duplicating someone is doing. By being able to draw from the function throughout a project is very powerful. Functions allow us to pick apart complex problems by breaking them into smaller pieces. Another advantage to creating a function is to improve the clairty of code. Functions are easy to read so when working on a project it is easier to look back to check on what has already been completed. Functions can easily be reused. 

Building all commands from scratch will waste someone's time by making the coding process incredibly inefficient.
