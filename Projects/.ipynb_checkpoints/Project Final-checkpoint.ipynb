## What is Malthusian Theory?

(Side note, I will be changing any description of man to human because there is no reason to use exclusive language)

Thomas Malthus was concerned with unregulated and overabundant population growth of humanity, "...to the life of man, population can never actually increase beyond the lowest nourishment capable of supporting it" (Malthus, 1967, p. 6). Land was the primary determinant of productivity before the first industrial revolution, people could only be as productive as the space they owned. When considering this mindset, it can be easy to conclude that productivity would have an upper limit and if productivity can not grow resources become more scarce as population increases. Population checks (Malthusian pressures) are either a preventative or positive check on population. Preventative checks on population are a voluntary, rational calculation where humans can predict distant consequences while positive checks occur in nature. Malthus predicts that humanity unchecked will cause economic consumption to decrease per person, "Will a person not lower their rank in life, and be obligated to give up in great measure to their former habits?" (Malthus, 1967, p. 12). Malthusian theory indicates that in a world of unchecked population growth, laborers will have to work more hours to get the same amount of pay they would receive in the past. Malthus outlines that when the real price of labor falls do to unregulated population growth, that there will be a wide inequality in wealth between the lower classes and farmers/capitalists. The structure of his work follows three principles: 1) Population is necessarily limited by the means of subsistence, 2) Population invariably increases where the means of subsistence increases unless prevented by some powerful and obvious checks (Malthusian pressures), 3) These checks, and the checks which repress the superior power of population, and keep its effects on a level with the means of subsistence are all resolvable into moral restraint, vice, and misery. 

# The Drawbacks of Malthusian Theory

Malthusian theory failed to incorporate technological improvements that increases (pushes out) a firm's production possibility frontier and with the development of new technologies we saw the production of new goods that previously weren't sold before. Malthusian theory implies that innovation and new markets are not impactful towards a representative agent’s consumption space with its focus on land being a constraint for productivity, "Different from the classical Malthusian theory, the two-sector framework allows technological shocks to change equilibrium living standards. A positive shock in luxury technology expands the production possibility frontier vertically" (Wu, 2015, p. 10). The creation of what is new, has been the biggest crux against Malthusian Theory. Thomas Robertson writes that the Malthusian description of poverty as a result of scarcity when applied to the United States is unrealistic, "...in part a discussion about poverty allows us to see that it was a story not only about humanity's relationship with nature but also about America's relations with the rest of the world, race and class relations, and family structure" (Robertson, 2012, p. 225). There are many determinants on why people are in the social economic status they are in and Malthusian theory is an irresponsible means of prescribing why people all over the world face poverty. Although the application of biological models has allowed people to place humans within nature and explain how we interact with biological forces, how we depend on raw material, and our evolutionary process there is the drawback that, "Malthusians often rigidly overemphasize the biological context of human actions to the exclusion of historical and cultural contexts" (Robertson, 2012, p. 228). This shows that Malthusian theory loses sight of the crucial ways that humans are not just another species. We have cultures and histories that are as complicated and diverse as any ecosystem. When we lose sight of our complicated political atmosphere we are unable realize how nature and culture blend together. Humans are thinking animals. We find solutions, "The result is the opposite of what Malthus predicted: the wealthiest nations with the greatest food security have the lowest fertility rates, whereas the most food-insecure countries have the highest fertility rates" (Shermer, 2016). In the modern era, the rate of population growth shows signs of decreasing thanks to the demographic transition. Wealthy nations like Japan and Europe face a new problem of declining populations and the United States would face this issue too if not for high levels of immigration. Laborers are not experiencing what Malthus predicted in his works, "Despite the massive population increase since then (the time of Malthus), the 'iron law of wages' has not depressed the incomes of the ordinary employee to the level of subsistence" (View, 2015). Overall Thomas Malthus was not able to describe real impacts that can be used in modern population and economic models. Modern economic models incorporate the complex social world around us and Malthusian theory is unable to consider the interconnections between humanity, the natural world, and culture while also failing to incorporate innovation and a functional labor market. 

## Data Structuring 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file = pd.ExcelFile("ds1.2022.xlsx")
file.sheet_names
```

```python
data = {}
for sheet_name in file.sheet_names:
    data[sheet_name] = pd.read_excel(file, sheet_name = sheet_name, parse_dates = True)
data
```

```python
data['GDP_1270-1700'] = data['GDP_1270-1700'].set_index(["Year"])

data['GDP_1270-1700'] = data['GDP_1270-1700'].set_axis(data["GDP_1270-1700"].iloc[0],
                    axis=1, inplace=False).iloc[2:]
data['GDP_1270-1700'].index = data['GDP_1270-1700'].index.astype(int)
data['GDP_1270-1700']
```

```python
data['GDP_1700-1870'].set_index("Year", inplace = True)
data['GDP_1700-1870'] = data['GDP_1700-1870'].set_axis(data['GDP_1700-1870'].iloc[0],
                    axis=1, inplace=False).iloc[2:]
data['GDP_1700-1870'].index = data['GDP_1700-1870'].index.astype(int)
data['GDP_1700-1870']
```

```python
keys = ["Agriculture", "Industry", "Services", "GDP"]
merge_data = data['GDP_1270-1700'].T.iloc[:4].T#[keys]
merge_data = merge_data.append(data['GDP_1700-1870'].T.iloc[:4].T.loc["1701":"1870"], ignore_index = False)
merge_data
```

```python
for key, val in merge_data.items():
    merge_data[key] = pd.to_numeric(val)
GDP = {}
GDP["Data"] = merge_data
GDP["Log"] = np.log(merge_data)
GDP["Rates"] = GDP["Log"].diff()
GDP["Rates"]
```

```python
GDPPC = {"Data":GDP["Data"].copy()}
pop = data["Population"].set_index("Year").rename(columns = {"Population of England, millions": "England",
                                                            "Population of Great Britain, millions":"Great Britain"})
GDPPC["Data"]
```

```python
for key in GDPPC["Data"].keys():
    GDPPC["Data"][key].loc[:"1700"] = GDPPC["Data"][key].loc[:"1700"].div(pop["England"].loc[:"1700"].values)
    GDPPC["Data"][key].loc["1701":] = GDPPC["Data"].loc["1701":][key].div(pop["Great Britain"].loc["1701":].values)
    
GDPPC["Data"]
```

```python
type(GDPPC)
```

```python
GDPPC["Log"] = np.log(GDPPC["Data"])
GDPPC["Rates"] = GDPPC["Log"].diff()
GDPPC["Rates"]
```

```python
def median(data):
    n = len(data)
    lst = sorted(data)
    if n % 2 != 0:
        middleindex = int((n - 1) / 2)
        median_ = lst[middleindex]
    else:
        uppermiddleindex = int(n / 2)
        lowermiddleindex = uppermiddleindex - 1 
        # pass slice with middle values to mean()
        median_ = mean(lst[lowermiddleindex: uppermiddleindex + 1])
    return median_
    
groupMedians = []
for x in yearGroups:
    groupMedians.append(median(x))
print(groupMedians)
```

```python
def variance(data, sample = True):
    listmean = mean(data)
    n = len(data)
    DOF = n - 1
    sumsquarediff = 0
    
    for val in data:
        diff = val - listmean
        sumsquarediff += (diff) ** 2
    if sample == False:
        variance_ = sumsquarediff / n
    else:
        variance_ = sumsquarediff / DOF
    return variance_

groupVariance = []
for x in yearGroups:
    groupVariance.append(variance(x, sample = True))
print(groupVariance)
```
