## Use a For Loop to call the first index in the Two Dimensional Array as $i$
## calling Two Dimensional Array[i] will yield the ith Row

```python
for i in range(len(two_dim_array)):
    print(i)
    print(two_dim_array[i])
```

## Passing the row to a second For Loop will cycle through the second
## (row) index as $j$ Calling row[j] will yield the jth value from the row

```python
print("i,j:")
for i in range(len(two_dim_array)):
    row = two_dim_array[i]
    for j in range(len(row)):
        print(i, j)
        print("val:", row[j])
```

```python
for row in two_dim_array:
    print(row)
    for val in row:
        print(val)
```

# Create Artificial Data

```python
macrodict = {"GDP": {},
             "Money": {},
             "Real GDP": {},
             "Price Level": {}}
for key in macrodict.keys():
    for i in range(1990, 2021):
        macrodict[key][i] = np.random.random() * 10000
macrodict
```

```python
macroDF = pd.DataFrame(macrodict)
macroDF
```

```python
macroDF[["GDP"]]
```

```python
macroDF["Velocity"] = macroDF["Price Level"].mul(macroDF["Real GDP"].div(macroDF["Money"]))
macroDF
```

## matplotlib

```python
import matplotlib.pyplot as plt
macroDF.plot()
```

```python
# plot using method method for matplotlib from the pandas dataframe
fig, ax = plt.subplots(figsize = (20, 12))
macroDF.plot(linewidth = 5, alpha = .7, ax = ax)
ax.legend(fontsize = 20)
```

```python
# our goal is to create a multiplot visualization with a seperate plot area for each variables
# count the number of variables
numvars = len(macroDF.keys())
# create a plot with a subplot for each variable (total 5)
fig, ax = plt.subplots(1, numvars, figsize = (20, 6))
# to plot each variable on each plot, cycle through index of the keys
for i in range(numvars):
    # call key by index number
    key = list(macroDF.keys())[i]
    color = "C" + str(i)
    alpha = .2 + .15 * i
    #plot key according to index number in ax
    macroDF[key].plot.line(color = color, ax = ax[i])
    ax[i].set_title(key,color = color, fontsize = 20)
```

## Plotting Income Distribution

```python
incomedata = pd.read_excel("https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls", header=[7])
incomedata = incomedata.rename(columns = {"Unnamed: 0": "Household Income"})
incomedata
```

```python
income = incomedata["Number"][1: -1].div(1000)
income.iloc[0] = 0
income.head()
```

```python
incomedata[["Household Income"]]
```

```python
# make a list that will serve as the index to replace current index
incomelevels = list(range(0, 201, 5))
incomelevels.append(250)
incomelevels = [str(val) + "k" for val in incomelevels]
incomelevels.append("250k+")
incomelevels
```

```python
incomedata = pd.DataFrame({"Income Level":np.array(incomelevels), "Households in Bracket":income})
incomedata["Cumulative"] = incomedata["Households in Bracket"].cumsum()
incomedata.set_index("Income Level", inplace = True)
incomedata.plot.line(secondary_y = "Cumulative")
incomedata
```

```python
totalhouseholds = incomedata["Households in Bracket"].sum().round(3)
incomedata["% in Bracket"] = incomedata["Households in Bracket"].div(totalhouseholds).mul(100)
incomedata["Cumulative %"] = incomedata["% in Bracket"].cumsum().round(3)
incomedata
```

```python
plt.rcParams.update({"font.size": 16})
plotDF = incomedata[["% in Bracket", "Cumulative %"]]
fig, ax = plt.subplots(figsize = (15,7))
plotDF.plot.bar(secondary_y = "Cumulative %", width = .95, alpha = .75, ax = ax)
x_ticklabels = ["$" + str(x) for x in plotDF.index]
ax.set_xticklabels(x_ticklabels)
y1_ticklables = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklables(y1_ticklables)
```

```python
# call the x labels
# ax.get_xticklabels()
newindex = ["$" + str(x) for x in plotDF.index]
# newindex = []
# for ix in plotDF.index:
#    newindex.append("$" + ix)
```

```python
#build a distribution of income with 25k bins instead of 5k bins
#start by counting by through rows by 5s in the orginal dataframe
income25kbin = incomedata.loc[::5]
#we were missing the final row with 250k+, so add this in the last row
#we created a new row using df.loc[newrowname]
income25kbin.loc["250k+"] = incomedata.iloc[-1]
#redefine the marginal bins by taking the difference of the cumulative bins
income25kbin["Households in Bracket"] = income25kbin["Cumulative"].diff().fillna(0)
# the information from the % in Bracket column was incorrect so we recalculate this using
# the new household
income25kbin["% in Bracket"] = income25kbin["Households in Bracket"].div(totalhouseholds).mul(100)
income25kbin
```

```python
plt.rcParams.update({"font.size": 20})
plotDF = income25kbin[["% in Bracket", "Cumulative %"]].iloc[1:]
fig, ax = plt.subplots(figsize = (15,10))
plotDF.plot.bar(secondary_y = "Cumulative %", width = .98, alpha = .75, ax = ax)
x_ticklabels = ["$" + str(x) for x in plotDF.index]
ax.set_xticklabels(x_ticklabels)
y1_ticklables = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklables(y1_ticklables)
```
